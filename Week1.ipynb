```python
from bs4 import BeautifulSoup ##A package to work with HTML data
import requests #A package to make HTTP requests
```

```python
link = "https://ic2s2-2023.org/program"
r = requests.get(link)
soup = BeautifulSoup(r.content)
```

Start by getting the names from the overview table

```python
table = soup.find("table",{"class":"tutorials"})
table_rows = table.find_all("tr")

#HERE I GET THE HEADER
ths = table_rows[0].find_all('th')
header = [th.text.replace("\n","") for th in ths]
    
#HERE I GET THE ROWS
rows = []
for tr in table_rows[1:]:
    tds = tr.find_all('td')
    for td in tds:
        td = td.text.replace("\n","")
        if "Keynote" in td:
            td = td.replace("Keynote - ","")
            rows.append(td)         
rows           
```

Now all the plenary and posters

```python
content = soup.find_all("ul",{"class":"nav_list"})
for nav in content:
    content_rows = nav.find_all("i")
    for c_row in content_rows:
        td = c_row.text.replace("\n","")
        for item in td.split(","):
            item = item.strip()
            if item not in rows:
                rows.append(item)
rows
```

## 2024

```python
import pandas as pd
```

```python
rows_2024 = []
oral = pd.read_excel("IC2S2_2024_oral_panels.xlsx")["Presentation authors"]
for row in oral:
    for line in row.split(";"):
        name = line.split(",")[0]
        name = name.strip()
        if name not in rows_2024:
            rows_2024.append(name)
len(rows_2024)
```

```python
poster = pd.read_excel("IC2S2_2024_posters.xlsx")["Poster authors"]
for row in poster:
    for line in row.split(";"):
        name = line.split(",")[0]
        name = name.strip()
        if name not in rows_2024:
            rows_2024.append(name)
len(rows_2024)
```

```python
talks = pd.read_excel("IC2S2_2024_lightning_talks.xlsx")["Presentation authors"]
for row in talks:
    for line in row.split(";"):
        name = line.split(",")[0]
        name = name.strip()
        if name not in rows_2024:
            rows_2024.append(name)
len(rows_2024)
```
